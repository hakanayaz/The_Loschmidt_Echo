Creating the qubits and naming.

```
q1, q2, q3, q4, q5 = cirq.LineQubit.range(5)
sqrt_iswap = cirq.ISWAP**0.5
#Creating the rand Rotation values for the RQC's
rand1 = np.random.random()
rand2 = np.random.random()
rand3 = np.random.random()
rand4 = np.random.random()
rand5 = np.random.random()
rand6 = np.random.random()
rand7 = np.random.random()
rand8 = np.random.random()
rand9 = np.random.random()
rand10 = np.random.random()
rand11 = np.random.random()
rand12 = np.random.random()
rand13 = np.random.random()
rand14 = np.random.random()
rand15 = np.random.random()
```

```
# Deep random quantum circuits (RQCs) build
operations = [cirq.Moment(cirq.H(q1), cirq.H(q2), cirq.H(q3), cirq.H(q4), cirq.H(q5)),
              cirq.Moment(sqrt_iswap(q1,q2),  sqrt_iswap(q3,q4)),
              cirq.Moment(sqrt_iswap(q2,q3), sqrt_iswap(q4,q5)), 
              cirq.Moment(cirq.rx(rand1).on(q1),cirq.ry(rand2).on(q3),cirq.rz(rand3).on(q2),
                          cirq.ry(rand4).on(q4),cirq.rx(rand5).on(q5)),
              cirq.Moment(sqrt_iswap(q2,q3), sqrt_iswap(q4,q5)),
              cirq.Moment(sqrt_iswap(q1,q2),  sqrt_iswap(q3,q4)),
              cirq.Moment(cirq.rx(rand6).on(q1),cirq.rz(rand7).on(q2),
                          cirq.ry(rand8).on(q3),cirq.rx(rand9).on(q4),cirq.ry(rand10).on(q5)),
              cirq.Moment(sqrt_iswap(q1,q2),  sqrt_iswap(q3,q4)),
              cirq.Moment(sqrt_iswap(q2,q3), sqrt_iswap(q4,q5)), 
              cirq.Moment(cirq.rx(rand11).on(q1),cirq.ry(rand12).on(q3),cirq.rz(rand13).on(q2),
                          cirq.ry(rand14).on(q4),cirq.rx(rand15).on(q5))]
# deneme = [cirq.Moment(cirq.H(q1), cirq.H(q2), cirq.H(q3), cirq.H(q4), cirq.H(q5)),
#               cirq.Moment(sqrt_iswap(q1,q2),  sqrt_iswap(q3,q4)),
#               cirq.Moment(sqrt_iswap(q2,q3), sqrt_iswap(q4,q5)), 
#               cirq.Moment(cirq.rx(rand1).on(q1),cirq.ry(rand2).on(q3),cirq.rz(rand3).on(q2),
#                           cirq.ry(rand4).on(q4),cirq.rx(rand5).on(q5)),
#               cirq.Moment(sqrt_iswap(q2,q3), sqrt_iswap(q4,q5)),
#               cirq.Moment(sqrt_iswap(q1,q2),  sqrt_iswap(q3,q4)),
#               cirq.Moment(cirq.rx(rand6).on(q1),cirq.rz(rand7).on(q2),
#                           cirq.ry(rand8).on(q3),cirq.rx(rand9).on(q4),cirq.ry(rand10).on(q5)),
#               cirq.Moment(sqrt_iswap(q1,q2),  sqrt_iswap(q3,q4), cirq.X(q5)),
#               cirq.Moment(cirq.X(q1),sqrt_iswap(q2,q3), sqrt_iswap(q4,q5)),
#               cirq.Moment(cirq.rx(rand11).on(q1),cirq.ry(rand12).on(q3),cirq.rz(rand13).on(q2),cirq.ry(rand14).on(q4),cirq.rx(rand15).on(q5))]

# deneme_circuit = cirq.Circuit(deneme)
# print(deneme_circuit)


# Defining the Circuit
mycircuit = cirq.Circuit(operations)
# Printing the Circuit
print(mycircuit)
print('---------------------------------------------------------------------------------------------')
# Creating another circuit with adding the inverse.
final_mycircuit = mycircuit + cirq.inverse(mycircuit) 
# Printing together (with Inverse)
print(final_mycircuit)

# final_deneme_circuit = deneme_circuit + cirq.inverse(mycircuit)
```

```
operations = [cirq.Moment(cirq.H(q1), cirq.H(q2), cirq.H(q3), cirq.H(q4), cirq.H(q5)),
              cirq.Moment(sqrt_iswap(q1,q2),  sqrt_iswap(q3,q4)),
              cirq.Moment(sqrt_iswap(q2,q3), sqrt_iswap(q4,q5)), 
              cirq.Moment(cirq.rx(rand1).on(q1),cirq.ry(rand2).on(q3),cirq.rz(rand3).on(q2),
                          cirq.ry(rand4).on(q4),cirq.rx(rand5).on(q5)),
              cirq.Moment(sqrt_iswap(q2,q3), sqrt_iswap(q4,q5)),
              cirq.Moment(sqrt_iswap(q1,q2),  sqrt_iswap(q3,q4)),
              cirq.Moment(cirq.rx(rand6).on(q1),cirq.rz(rand7).on(q2),
                          cirq.ry(rand8).on(q3),cirq.rx(rand9).on(q4),cirq.ry(rand10).on(q5)),
              cirq.Moment(sqrt_iswap(q1,q2),  sqrt_iswap(q3,q4)),
              cirq.Moment(sqrt_iswap(q2,q3), sqrt_iswap(q4,q5)), 
              cirq.Moment(cirq.rx(rand11).on(q1),cirq.ry(rand12).on(q3),cirq.rz(rand13).on(q2),
                          cirq.ry(rand14).on(q4),cirq.rx(rand15).on(q5))]
possibilities = [cirq.Moment(cirq.H(q1), cirq.H(q2), cirq.H(q3), cirq.H(q4), cirq.H(q5)),
                 cirq.Moment(sqrt_iswap(q1,q2),  sqrt_iswap(q3,q4), cirq.X(q5)),
                 cirq.Moment(cirq.X(q1),sqrt_iswap(q2,q3), sqrt_iswap(q4,q5)), 
                 cirq.Moment(cirq.rx(rand1).on(q1),cirq.ry(rand2).on(q3),cirq.rz(rand3).on(q2),
                          cirq.ry(rand4).on(q4),cirq.rx(rand5).on(q5)),
                 cirq.Moment(cirq.X(q1),sqrt_iswap(q2,q3), sqrt_iswap(q4,q5)), 
                 cirq.Moment(sqrt_iswap(q1,q2),  sqrt_iswap(q3,q4), cirq.X(q5)),
                 cirq.Moment(cirq.rx(rand6).on(q1),cirq.rz(rand7).on(q2),
                          cirq.ry(rand8).on(q3),cirq.rx(rand9).on(q4),cirq.ry(rand10).on(q5)),
                 cirq.Moment(sqrt_iswap(q1,q2),  sqrt_iswap(q3,q4), cirq.X(q5)),
                 cirq.Moment(cirq.X(q1),sqrt_iswap(q2,q3), sqrt_iswap(q4,q5))]
# Creating the random numbers for putting the X gate.
j = np.random.randint(1,9)
# Because of there is no place to put X gate in that specifik moments.
if j ==3 or j==6:
  j = np.random.randint(1,6)
print('Random moment place is: ',j)

new_opt = operations
new_opt[j] = possibilities[j]
# Printing the Rand X location and inverse of the orginal circuit.
my_circuit_added_rand_X = cirq.Circuit(new_opt) + cirq.inverse(mycircuit)
print(my_circuit_added_rand_X)
```
